```python
import pandas as pf
```

```python
import numpy as np
```

```python
pf = pf.read_csv('/content/logistics_dataset_uncleaned.csv')
```

```python
pf.drop_duplicates(subset=['Shipment_ID'],inplace=True)
```

```python
pf.drop_duplicates(subset=['Order_ID'],inplace=True)
```

```python
pf.drop_duplicates(subset=['Customer_ID','Email'],inplace=True)
```

```python
pf['City'].fillna('Unknown',inplace=True)
```

```python
pf['Country'].fillna('unKnown',inplace=True)
```

```python
pf['Category'].fillna('Misc',inplace=True)
```

```python
pf['Weight_kg'].fillna(pf['Weight_kg'].mean())
```

```python
pf['Delivery_Mode'].fillna('Road',inplace=True)
```

```python
pf['Fuel_Cost']=pf.groupby('Transport_Type')['Fuel_Cost'].transform(lambda x: x.fillna(x.median()))
```

```python
pf['Remarks'].fillna('No Remarks',inplace=True)
```

```python
pf= pf[pf['Quantity'] > 0]
```

```python
pf=pf[pf['Weight_kg'] > 0]
```

```python
pf=pf[pf['Total_Price'] > 0]
```

```python
pf=pf[pf['Delivery_Date'] >= pf['Shipment_Date']]
```

```python
pf=pf[pf['Delay_Days'] >= -1]
```

```python
pf['Payment_Mode'] = pf['Payment_Mode'].replace({'Credt':'Credit','Debet':'Debit'})
```

```python
pf['Status']= pf['Status'].str.title
```

```python
pf['Country']=pf['Country'].replace({'us':'USA','U.K':'UK'})
```

```python
pf=pf[pf['Email'].str.contains('@',na=False)]
```

```python
pf=pf[pf['Distance_km'] <= 2000 ]
```

```python
pf['total_value_per_customer'] = pf.groupby("Customer_ID")['Total_Price'].sum()
```

```python
pf['avg_fuel']= pf.groupby("Delivery_Mode")['Fuel_Cost'].mean()
```

```python
warehouses_multi_regions = pf.groupby("Warehouse_Name")['Region'].nunique()
warehouses_multi_regions = warehouses_multi_regions[warehouses_multi_regions > 1]
```

```python
pf['Revenue'] = pf['Quantity'] * pf['Unit_Price']
```

```python
pf['Fuel_Efficiency'] = pf['Distance_km'] / pf['Fuel_Cost']
```

```python
pf['Profit'] = pf['Total_Price'] - pf['Fuel_Cost']
```

```python
pf['Delay_Flag'] = np.where(pf['Delay_Days'] > 0, "Yes", "No")
```

```python
pf['Urgency_Level'] = pf['Order_Priority'].map({'High':'Urgent','Medium':'Normal','Low':'Low'})
```

```python
pf['Payment_Mode_Encoded'] = pf['Payment_Mode'].astype('category').cat.codes
```

```python
pf['Order_Priority_Encoded'] = pf['Order_Priority'].map({'High':3,'Medium':2,'Low':1})
```

```python
top5_products = pf.groupby('Product_ID')['Revenue'].sum().nlargest(5)
```

```python
top_driver = pf['Driver_Name'].value_counts().idxmax()
```
